【PythonとStanで学ぶ】仕組みが分かるベイズ統計学入門

本コースは、名前の通りベイズ統計学の入門コースです。近年、機械学習が世間を賑わせていますが、ベイズ統計学もまた近年非常に着目されている統計学です。

このベイズ統計学は従来の統計学とはアプローチが少し異なり、観測されたデータから、逆に原因であるパラメータを確率分布として推定を行うという帰納的アプローチを取る統計学です。この為、昨今のデータサイエンスと相性がよく、機械学習には所謂、過学習の問題がありますが、ベイズ統計学のアプローチでは、データが少なければ、信頼区間が広がるだけでデータのスケールに応じた不確定性を推定するだけです。

つまり、データに対するスケーラビリティがあり、過学習しにくいです。

このように、非常にデータサイエンスとしては魅力的なベイズ統計学ですが、実際にビジネスの現場で使おうとするとハードルがあります。

それは、非常に他分野に渡る高度な知識を要求されるという事です。従来の統計学の知識はもちろんのこと、確率分布に関する知識、ベイズ統計学の知識、MCMCというベイズ統計学では必須となってくるアルゴリズムへの理解。またMCMCの原理として、例えば、ハミルトニアンモンテカルロなどを理解したい場合には、物理的な知識を必要とします。

この為、専門書も数学的に高度になりがちで、ある意味それが、ベイズ統計学普及の妨げになっており、ベイズ統計学最大の欠点だといえます。

本コースでは、そのような様々な障壁を乗り越えるために、確率の基礎から始まって、ベイズ統計学の基礎、MCMCのアルゴリズムの理解や数値計算の知識、さらには、MCMCのフレームワークであるStanの使い方など、ベイズ統計学に必要な知識を網羅的・体系的に纏めています。また、付録には簡単ではありませんが、ハミルトニアンモンテカルロを理解するための初歩的な物理学の基礎の解説も行っています。

やはり、広範な知識と比較的高度な知識を要求されるので、講義のボリュームとしては11時間ありますが、実際のところ、付録までしっかりやろうとすると、15時間くらい必要だと思います。ちなみにこれでも十分な量ではありません。ボーナスレクチャーには参考文献を付けていますので、本コースを学んだ後に、それらの参考文献を紐解く+α自助努力して頂くことで、論文を書けるレベルには到達できると思います。
